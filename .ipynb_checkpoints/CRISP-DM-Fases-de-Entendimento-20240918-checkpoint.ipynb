# Fase 1: Entendimento do Negócio

The **Business Understanding** phase focuses on understanding the objectives and requirements of the project.

1. **Determine business objectives**: You should first “thoroughly understand, from a business perspective, what the customer really wants to accomplish.” and then define business success criteria.
2. **Assess situation**: Determine resources availability, project requirements, assess risks and contingencies, and conduct a cost-benefit analysis.
3. **Determine data mining goals**: In addition to defining the business objectives, you should also define what success looks like from a technical data mining perspective.
4. **Produce project plan**: Select technologies and tools and define detailed plans for each project phase.

Establishing a strong business understanding is like building the foundation of a house – absolutely essential.

### Funções Estatísticas

#### Sumarização

```python
def value_counts(series):
    """ Calcula frequências dos valores de dada série.
    """
    counts_abs = series.value_counts()
    counts_rel = series.value_counts(normalize=True)
    return pd.DataFrame({
        'abs': counts_abs, 
        'rel': counts_rel,
        'cum_abs': counts_abs.cumsum(),
        'cum_rel': counts_rel.cumsum()
    })

def IQR(df):
    """ Calcula distância interquartil.
    """
    return df.quantile(.75) - df.quantile(.25)
```

#### Amostragem

```python
def sample_size(conf=.95, erro=.031, N=None):
    """ Determina tamanho da amostra.
    
    Parâmetros:  
        conf : float 
            Nível de confiança desejado, 0 < conf < 1.
        erro : float 
            Margem de erro desejada,     0 < erro < 1.
            
        N : int 
            Tamanho da população.
        
    Retorna:
        n : int
            Tamanho da amostra. Se N é passado, calcula tamanho da
            amostra de modo heurístico. Senão, calcula tamanha da
            amostra a partir do intervalo de confiança desejado.
    """
 
    if N:
        n = max(100, min(N * 0.1, 10000))
    else:
        if       conf <= .80: z = 1.28
        if .80 < conf <= .90: z = 1.65
        if .90 < conf <= .95: z = 1.96
        if .95 < conf       : z = 2.58
        n = np.ceil(0.25*(z/erro)**2)
    return int(n)
```

#### Preparação de Dados

```python
# max absolute scaling 
def maxabs_scaling(df):
    return df/df.abs().max()

# max-min scaling
def maxmin_scaling(df):
    return (df-df.min())/(df.max()-df.min())

# z-score scaling
def zscore_scaling(df):
    return (df-df.mean())/df.std(ddof=0)

# IQR based scaling 
def robust_scaling(df):
    IQR = df.quantile(.75) - df.quantile(.25)
    return (df-df.median())/IQR


def value_scalings(series):
    """ Normaliza valores; Métodos: maxabs, maxmin, zscore e IQR.
    """
    return pd.DataFrame({
        'value': series,
        'maxabs': maxabs_scaling(series),
        'maxmin': maxmin_scaling(series),
        'zscore': zscore_scaling(series),
        'robust': robust_scaling(series)
    })


# anota linhas com outliers:
def detect_outliers(df, columns='all', zmax=3, zscore=robust_scaling, outcol='outlier'):
    if columns == 'all':
        columns = df.columns
    elif isinstance(columns, str):
        columns = [columns]
        
    df[outcol] = (zscore(df[columns]).abs() > zmax).any(axis=1)
    return df 
```
